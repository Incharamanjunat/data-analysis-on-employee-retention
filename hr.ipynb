```python
# Data Preprocessing.

#Identify and treating of missing values.

data.isnull().sum() # Countsthe number of missing values in each column of the DataFrame 'data'
```

```python
data['EducationField'].fillna("Bachelor's", inplace=True) # Fills missing values in the 'EducationField' column of the DataFrame 'data' with "Bachelor's"
```

```python
data['Attrition'].skew # Calculates the skewness of the 'Attrition' column in the DataFrame 'data'
```

```python
data['EnvironmentSatisfaction'].fillna(data['EnvironmentSatisfaction'].median(),                                        
                                       inplace=True)  # Fills missing values in the 'EnvironmentSatisfaction' column of the DataFrame 'data' with the median value of the same column
```

```python
# Check duplicate records.

data[data.duplicated()]
```

```python
data = data.drop_duplicates() # Removes duplicate rows from the DataFrame 'data'
```

```python
data.shape # Returns the dimensions (number of rows, number of columns)
```

```python
data.head(10)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder() # LabelEncoder is a class used to transform categorical labels into numerical labels
```

```python
data['Department'] = le.fit_transform(data['Department'])
data['Gender'] = le.fit_transform(data['Gender'])
data['BusinessTravel'] = le.fit_transform(data['Education'])
data['EducationField'] = le.fit_transform(data['EducationField']) # Using LabelEncoder to transform categorical columns in the DataFrame 'data' ('Department', 'Gender', 'BusinessTravel', 'EducationField') into numerical labels
```

```python
data.drop(labels='EmployeeNumber',axis=1,inplace=True) # Drops the column 'EmployeeNumber' from the DataFrame 'data'
```

```python
data.head(3)
```

```python
data.columns
```

```python
rel_feat = ['Age', 'Attrition', 'BusinessTravel', 'DailyRate', 'Department',
       'DistanceFromHome', 'Education', 'EducationField', 'EmployeeCount',
       'EnvironmentSatisfaction', 'Gender', 'HourlyRate', 'JobInvolvement',
       'JobLevel', 'JobRole', 'JobSatisfaction', 'MaritalStatus']
```

```python
# Selecting specific columns from the DataFrame 'data'
selected_columns = ['Age', 'Attrition', 'BusinessTravel', 'DailyRate', 'Department',
       'DistanceFromHome', 'Education', 'EducationField', 'EmployeeCount',
       'EnvironmentSatisfaction', 'Gender', 'HourlyRate', 'JobInvolvement',
       'JobLevel', 'JobRole', 'JobSatisfaction', 'MaritalStatus']

# Creating a subset DataFrame with selected columns
subset_data = data[selected_columns]

# Encoding categorical variables using one-hot encoding
subset_data_encoded = pd.get_dummies(subset_data, columns=['Attrition', 'BusinessTravel', 'Department', 'EducationField', 'Gender', 'JobRole', 'MaritalStatus'])
# Converts categorical variables into dummy/indicator variables using one-hot encoding, replacing each categorical column specified in 'columns' with binary indicator columns

# Calculating the correlation of each feature with 'JobSatisfaction'
rel_feat_corr = subset_data_encoded.corr()['JobSatisfaction']

print(rel_feat_corr)
```

```python
# Selects numeric features related to a specific target variable ('rel_feat'), computes their correlation matrix, and visualizes it using a heatmap
numeric_rel_feat = data[rel_feat].select_dtypes(include=['number'])

plt.figure(figsize=(12, 12))
sns.heatmap(numeric_rel_feat.corr(), annot=True, linewidths=0.5)
plt.show()
```

```python
x = data[rel_feat]
y = data['JobSatisfaction']
```

```python
from sklearn.model_selection import train_test_split
x_data, x_test, y_data, y_test = train_test_split(x, y, test_size=0.2, random_state=0) # Splits the dataset into training and testing sets using train_test_split, with 20% of the data reserved for testing and a fixed random state
```

```python
# Creates a histogram to visualize the distribution of ages for two groups based on 'JobSatisfaction' (1 and 0) in the DataFrame 'data'. Sets the figure size to 6x4 inches.

figure = plt.figure(figsize=(6,4))
plt.hist([data[data['JobSatisfaction'] == 1]['Age'],
         data[data['JobSatisfaction'] == 0]['Age']],
        color = ['g','r'],
        bins = 10,label = ['1','2'])
plt.xlabel('Age')
plt.ylabel('# EmployeeNumber')
plt.legend()
```

```python
# Adds a new column 'Department_codes' to the DataFrame 'data' representing the categorical codes of the 'Department' column. 

data['Department_codes'] = data['Department'].astype('category').cat.codes
# Creates a scatter plot to visualize the relationship between 'Age' and 'HourlyRate', with different colors representing different departments based on 'Department_codes'. 
# Sets the colormap to 'jet', adjusts the transparency (alpha) of the points to 0.8, and sets the figure size to 8x5 inches.
data.plot.scatter('Age', 'HourlyRate', c='Department_codes', colormap='jet', alpha=0.8, figsize=(8, 5))

plt.show()
```

```python
# Creates a box plot to visualize the distribution of 'HourlyRate' across different levels of 'EducationField', with separate boxes for each level of 'JobSatisfaction'. 
# Adjusts the figure size to 9x6 inches and sets the Seaborn style to 'darkgrid'. 
# Places the legend at the best location.

plt.figure(figsize=(9,6))
sns.set_style('darkgrid')
sns.boxplot(x='EducationField',y='HourlyRate',data=data, hue="JobSatisfaction")
plt.legend(loc=0)
```

```python
# Creates a strip plot to visualize the distribution of 'HourlyRate' across different levels of 'EducationField', with separate points for each level of 'JobSatisfaction'. 
# Adjusts the figure size to 8x5 inches and adds some random noise to the points along the categorical axis ('EducationField') using the jitter parameter. 
# Separates the points for each level of 'JobSatisfaction' and avoids overlapping using the hue parameter and dodge=True.

plt.figure(figsize=(8,5))
sns.stripplot(x='EducationField',y='HourlyRate',data=data,jitter=True,hue='JobSatisfaction',
             dodge=True)
```
